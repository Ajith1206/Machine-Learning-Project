## Now going to apply condition for column ("loan_status") by ("last_fico_range_high") following :
For Fully Paid states whose last_fico_range_high is above 579 will be in "Pay" rest all goes under ""Not Pay"" statues.

For Current states whose last_fico_range_high is above 579 will be in "Pay" rest all goes under ""Not Pay"" statues.

For Does not meet the credit policy. Status:Fully Paid states whose last_fico_range_high is above 579 will be in "Pay" rest all goes under ""Not Pay"" statues.

For Does not meet the credit policy. Status:Charged Off states whose last_fico_range_high is above 799 will be in "Pay" rest all goes under ""Not Pay"" statues.

Rest all columns given diectly to "Not Pay" status.

```python
def fill_age(cols):                              # created function condition to fit "pay" or" not pay" 
    loan_status = cols[0]
    last_fico_range_high = cols[1]
    
    if (loan_status=="Fully Paid"):
        if last_fico_range_high > 579:
            return "Pay"
        else:
            return "Not Pay"
    else:
        return loan_status
```

# Detecting Non Outleirs Count

```python
def remove_outlier_count(col_name):             #cretaed function to detect outlier
    q1 = df2[col_name].quantile(0.25)
    q3 = df2[col_name].quantile(0.75)
    iqr = q3-q1 #Interquartile range
    lower_Limit = q1-1.5*iqr
    upper_Limit = q3+1.5*iqr
    df_updated = df2.loc[(df2[col_name] > lower_Limit) & (df2[col_name] < upper_Limit)]
    return len(df_updated)
```
